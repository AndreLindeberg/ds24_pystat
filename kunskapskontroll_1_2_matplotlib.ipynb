### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# OO-style (object-oriented style), wich is recomended. Explicit
# Pyplot-style, convenient for quick interactive work. Implicit 
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# figure, is the whole figure that keeps the track of all information. Such as axes, titles and data displayed.
# axes, is the area we are plotting, making our digaram or graf. It contains the lables of x, y and if used z-axes. Also all the elements containing the graf. Such as, titles and ticklables.
# axis, are referring to the x, y and z-axels. The lines or ticks representing the coordinates. 
# artist, is all the visual elements in the figure.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
# numpy.array or numpy.ma.masked_array. Or objects that can be passed as numpy.asarray.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100) 

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x**2, label='test')
ax.set_xlabel('x lable')
ax.set_ylabel('y lable')
ax.set_title('Simple Plot')
ax.legend()
```

```python
x = np.linspace(-4, 4, 100) 

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x**2, label='test')
plt.xlabel('x lable')
plt.ylabel('y lable')
plt.title('My Plot')
plt.legend()
```

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter(random_data_x, random_data_y)
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('My scatter')
```

```python
# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(names, values)
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('My Plot')
```
